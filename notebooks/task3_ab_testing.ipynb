Import libraries


```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind, mannwhitneyu, chi2_contingency
from statsmodels.stats.proportion import proportions_ztest
```

Load your dataset

```python
df = pd.read_csv(r"C:\Users\mihre\Documents\KIAM\Project week3\acis_insurance_risk\acis_insurance_risk\data\processed\cleaned_policies.csv")


# Ensure required fields exist
required_cols = ["Segment", "claim_flag", "severity", "margin"]
for col in required_cols:
    if col not in df.columns:
        raise ValueError(f"Missing required column: {col}")

```

Define groups

```python
group_A = df[df["Segment"] == "A"]
group_B = df[df["Segment"] == "B"]

print(f"Group A size: {len(group_A)}")
print(f"Group B size: {len(group_B)}")
```

Hypothesis Testing

Claim Frequency — Proportion Test

```python
from statsmodels.stats.proportion import proportions_ztest

count = [group_A.claim_flag.sum(), group_B.claim_flag.sum()]
nobs = [len(group_A), len(group_B)]
stat_freq, p_freq = proportions_ztest(count, nobs)
stat, p_value = proportions_ztest(count, nobs)
print(stat, p_value)
```

Claim Severity — T-test or Mann–Whitney

```python
severity_A = group_A.severity.dropna()
severity_B = group_B.severity.dropna()

ttest_stat, ttest_p = ttest_ind(severity_A, severity_B, equal_var=False)
mw_stat, mw_p = mannwhitneyu(severity_A, severity_B, alternative="two-sided")
```

Margin Difference

```python
margin_A = group_A.margin.dropna()
margin_B = group_B.margin.dropna()

margin_stat, margin_p = ttest_ind(margin_A, margin_B, equal_var=False)
```

```python
# ------------------------------------------------------------
# Print Summary
# ------------------------------------------------------------


summary = f"""
===============================
     A/B Hypothesis Testing
===============================

Claim Frequency:
- Z-statistic: {stat_freq:.4f}
- p-value: {p_freq:.4f}

Claim Severity:
- T-test p-value: {ttest_p:.4f}
- Mann–Whitney p-value: {mw_p:.4f}

Margin Difference:
- T-test p-value: {margin_p:.4f}

Conclusion:
- If p < 0.05 → statistically significant difference
===============================
"""
print(summary)

import os

# Ensure the directory exists
output_dir = "models/evaluation_reports"
os.makedirs(output_dir, exist_ok=True)

# Save summary with UTF-8 encoding
with open(os.path.join(output_dir, "task3_ab_test_summary.txt"), "w", encoding="utf-8") as f:
    f.write(summary)

print(f"Task 3 summary saved to {output_dir}/task3_ab_test_summary.txt")
```
